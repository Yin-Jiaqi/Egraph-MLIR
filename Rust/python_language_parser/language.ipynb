```python
import re

def convert_pattern(input_string):
    print(input_string)
    # Parse the input string using regex
    pattern = r'"(.*?)" = (.*?)\(\[Id; (\d+)\]\)'
    matches = re.findall(pattern, input_string)

    result = []
    for match in matches:
        print("111")
        key, function, size = match
        size = int(size)
        
        # Format the desired pattern
        arr_content = ", ".join([f"in_id[{i}]" for i in range(size)])
        formatted_str = f'"{key}" => {{\n    let arr: [Id; {size}] = [{arr_content}];\n    Some(SimpleLanguage::{function}(arr))\n}},'
        result.append(formatted_str)

    return "\n".join(result)

# Reading from input file
with open('language.txt', 'r') as infile:
    input_string = infile.read()

# Converting the pattern
output_string = convert_pattern(input_string)

# Writing to the output file
with open('language_enum.txt', 'w') as outfile:
    outfile.write(output_string)
```
